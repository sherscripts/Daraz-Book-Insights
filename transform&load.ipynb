```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
daraz_df=pd.read_csv("raw_data.csv")
```

```python
daraz_df.head(10)
```

```python
null_values=daraz_df.isnull().sum()
print(null_values)
```

```python
daraz_df['price']=daraz_df['price'].str.replace('Rs. ','').str.replace(',','').astype(float)
```

```python
daraz_df.head()
```

```python
def clean_unitsold(unit_sold):
    try:
        unit_sold=unit_sold.replace(' sold','')
        if 'K' in unit_sold:
          return float(unit_sold.replace('K',''))*1000
        return float(unit_sold)
    except:
        return 0

    
daraz_df['unit_sold']=daraz_df['unit_sold'].apply(clean_unitsold)
```

```python
daraz_df.head(50)
```

```python
daraz_df.dtypes
```

```python
daraz_df['rating_count']=daraz_df['rating_count'].astype(int)
```

```python
daraz_df.duplicated().sum()
```

```python
daraz_df['author'] = daraz_df['title'].str.extract(r'By (.+)')
```

```python
daraz_df['author']=daraz_df['author'].fillna('Unknown')
```

```python
daraz_df.head(100)
```

```python
daraz_df.describe()
daraz_df[(daraz_df['unit_sold'] == 0) & (daraz_df['stars'] >= 4)]
```

```python
daraz_df=daraz_df[~((daraz_df['unit_sold']==0)&(daraz_df['rating_count']<=1))]
```

```python
daraz_df[(daraz_df['unit_sold'] == 0) & (daraz_df['stars'] >= 4)]
```

```python
daraz_df.to_csv('books_data.csv',index=False)
```

```python
top_sellers = daraz_df.sort_values("unit_sold", ascending=False).head(10)

plt.figure(figsize=(10, 10))
sns.barplot(data=top_sellers, x="unit_sold", y="title", palette="Blues_r")
plt.title("Most Popular Books (by Units Sold)")
plt.xlabel("Units Sold")
plt.ylabel("Book Title")
plt.tight_layout()
plt.show()
```

```python
hidden_gems = daraz_df[(daraz_df["stars"] >= 4.5) & (daraz_df["rating_count"] < 10) & (daraz_df["unit_sold"] < 50)]

gem_list = hidden_gems[["title", "author", "price", "stars", "unit_sold", "rating_count"]]
print(gem_list)
```
